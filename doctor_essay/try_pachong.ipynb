```python

url = r'http://sousuo.gov.cn/s.htm?t=zhengce&q=发改委'
res = urllib.urlopen(url)
html = res.read().decode('utf-8')
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
result = soup.find_all('div')
```
