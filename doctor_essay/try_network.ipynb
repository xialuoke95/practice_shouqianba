```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.font_manager import _rebuild
_rebuild()
```

```python
G = nx.Graph()#创建空的网络图
G = nx.DiGraph()
G = nx.MultiGraph()
G = nx.MultiDiGraph()

G.add_node('a')#添加点a
G.add_edge('x','y')#添加边,起点为x，终点为y
```

```python
G = nx.petersen_graph()
plt.subplot(121)

nx.draw(G, with_labels=True, font_weight='bold')
plt.subplot(122)
```

```python
G.edges
```

```python
for _ in G.edges:
    a,b = _
    G[a][b]['weight'] = 1.0

G[1][2]['weight'] = 2.0
```

```python
options = {
    "node_color": "#A0CBE2",
    "edge_color": range(15),
    "width": [G[a][b]['weight'] for a,b in G.edges],
    "edge_cmap": plt.cm.Blues,
}
# nx.draw_shell(G,  nlist=[range(7,10), range(4, 7), range(4)], with_labels=True, font_weight='bold',**options)
nx.draw_shell(G,  nlist=[range(5,10), range(5)], with_labels=True, font_weight='bold',**options)
```
